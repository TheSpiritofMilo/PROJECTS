## Getting insights using SQL

### Question 1. 
Who won the Oscar for “Actor in a Leading Role” in  2015?

```python
%%sql 
SELECT * FROM oscars
WHERE year = 2015
```

### Question 2

What are the ten oldest movies in the database?

```python
%%sql
SELECT * FROM movies WHERE release_date IS NOT NULL ORDER BY release_date ASC LIMIT 10
```

### Question 3

How many unique awards are there in the Oscars table?

```python
%%sql
SELECT COUNT(DISTINCT award) as NUMBER
FROM oscars
```

### Question 4

How many movies are there that contain the word “Spider” within their title?



```python
%%sql
SELECT 
COUNT(original_title)
FROM movies
WHERE original_title LIKE '%Spider%'
```

### Question 5

How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?

```python
%%sql
SELECT 
COUNT(DISTINCT A.movie_id)
FROM movies AS A
JOIN genremap AS B
ON A.movie_id = B.movie_id
JOIN genres AS C
ON C.genre_id = B.genre_id
JOIN keywordmap AS D
ON A.movie_id = D.movie_id
JOIN keywords AS E
ON D.keyword_id = E.Keyword_id
WHERE E.keyword_name LIKE '%love%'
AND C.genre_name LIKE 'Thriller'
```

### Question 6

How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?

```python
%%sql
SELECT 
COUNT(title)
FROM movies
WHERE popularity > 40
AND budget < 50000000
AND release_date > '2006-08-01'
AND release_date < '2009-10-01'
```

### Question 7

How many unique characters has "Vin Diesel" played so far in the database?

```python
%%sql
SELECT COUNT(characters)
FROM actors AS A
JOIN casts AS B
ON A.actor_id = B.actor_id
WHERE actor_name LIKE 'Vin Diesel'
```

### Question 8

What are the genres of the movie “The Royal Tenenbaums”?

```python
%%sql
SELECT 
C.genre_name
FROM
movies AS A
JOIN genremap AS B
ON A.movie_id = B.movie_id
JOIN genres AS C
ON B.genre_id = C.genre_id
WHERE title LIKE 'The Royal Tenenbaums'
```

### Question 9

What are the three production companies that have the highest movie popularity score on average, as recorded within the database?

```python
%%sql
SELECT 
production_company_name,
AVG(A.popularity) 
FROM
movies AS A
JOIN productioncompanymap AS B
ON A.movie_id = B.movie_id
JOIN productioncompanies AS C
ON B.production_company_id = C.production_company_id
GROUP BY production_company_name
ORDER BY (AVG(A.popularity)) DESC
LIMIT 5

```

```python
%%sql
SELECT 
production_company_name,
(A.popularity) 
FROM
movies AS A
JOIN productioncompanymap AS B
ON A.movie_id = B.movie_id
JOIN productioncompanies AS C
ON B.production_company_id = C.production_company_id
WHERE production_company_name = "MCL Films S.A."
LIMIT 5

```

### Question 10

How many female actors (i.e. gender = 1) have a name that starts with the letter "N"?

```python
%%sql
SELECT COUNT(actor_name)
FROM actors
WHERE gender = 1
AND actor_name LIKE 'N%'
```

### Question 11

Which genre has, on average, the lowest movie popularity score? 

```python
%%sql
SELECT 
C.genre_name,
AVG(Popularity)
FROM
movies AS A
JOIN genremap AS B
ON A.movie_id = B.movie_id
JOIN genres AS C
ON B.genre_id = C.genre_id
GROUP BY C.genre_name
ORDER BY AVG(Popularity) ASC
LIMIT 5
```

### Question 12

Which award category has the highest number of actor nominations (both male and female actors)?

```python
%%sql
SELECT 
award,
COUNT(name)
FROM Oscars
WHERE award LIKE '%actor%'
OR award LIKE '%actress%'
GROUP BY award
ORDER BY COUNT(name) DESC
```
